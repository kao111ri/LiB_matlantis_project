```python
import pandas as pd
import numpy as np

# ############################################################################
# 
# ############################################################################

def calculate_interfacial_density(df):
    """
    3
    (substrate)
    """
    # --- 1:  ---
    area_map = {
        'Al2O3': 13.94673902 * 13.08364686,
        'AlF3':  11.76094577 * 12.80689447,
        'Al':    16.952258347397713 * 16.952258350558857
    }
    df['cross_sectional_area_A2'] = df['substrate'].map(area_map)

    if df['cross_sectional_area_A2'].isnull().any():
        unknown_substrates = df[df['cross_sectional_area_A2'].isnull()]['substrate'].unique()
        print(f": : {unknown_substrates}")
        print("0 area_map ")
        df['cross_sectional_area_A2'].fillna(0, inplace=True)

    # --- 2:  ---
    # 
    substrate_elements = ['Al']
    df['substrate_mixed_atoms'] = 0
    for elem in substrate_elements:
        col = f'{elem}_upper'
        if col in df.columns:
            df['substrate_mixed_atoms'] += df[col].fillna(0)

    # 
    cathode_elements = ['Li', 'Mn', 'Co', 'Ni']
    df['cathode_mixed_atoms'] = 0
    for elem in cathode_elements:
        col = f'{elem}_lower'
        if col in df.columns:
            df['cathode_mixed_atoms'] += df[col].fillna(0)

    # --- 3:  ---
    valid_area_mask = df['cross_sectional_area_A2'] > 0
    
    # 0
    df['substrate_contamination_density'] = 0.0
    df['cathode_contamination_density'] = 0.0
    
    # 
    df.loc[valid_area_mask, 'substrate_contamination_density'] = df.loc[valid_area_mask, 'substrate_mixed_atoms'] / df.loc[valid_area_mask, 'cross_sectional_area_A2']
    # 
    df.loc[valid_area_mask, 'cathode_contamination_density'] = df.loc[valid_area_mask, 'cathode_mixed_atoms'] / df.loc[valid_area_mask, 'cross_sectional_area_A2']
    #  ()
    df['interfacial_contamination_density'] = df['substrate_contamination_density'] + df['cathode_contamination_density']

    return df

def normalize_metric(series, high_is_good=True):
    """
    PandasSeries
    """
    series_finite = series.replace([np.inf, -np.inf], np.nan).fillna(0)
    min_val = series_finite.min()
    max_val = series_finite.max()
    
    if max_val - min_val == 0:
        return pd.Series(0.5, index=series.index)

    if high_is_good:
        return (series_finite - min_val) / (max_val - min_val)
    else:
        return (max_val - series_finite) / (max_val - min_val)

def score_for_clean_separation(csv_path="/home/jovyan/Kaori/MD/LiB_2/output/comprehensive_analysis_output/comprehensive_analysis_results_3.csv"):
    """
    
    """
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f":  '{csv_path}'")
        return

    print("...")

    # ---  ---
    metrics_config = {
        'tensile_strength_GPa':              {'weight': 0.2, 'high_is_good': False},
        'interfacial_contamination_density': {'weight': 0.8, 'high_is_good': False},
    }

    # 
    df = calculate_interfacial_density(df)

    # ---  ---
    df[''] = 0.0
    
    print("\n:")
    for metric, config in metrics_config.items():
        direction = "" if not config['high_is_good'] else ""
        print(f"- {metric}:  {config['weight']*100:.0f}%, ({direction})")
        
        if metric not in df.columns:
            print(f"  :  '{metric}' ")
            continue
            
        df[metric].fillna(0, inplace=True)
        
        normalized_col_name = f'norm_{metric}'
        df[normalized_col_name] = normalize_metric(df[metric], high_is_good=config['high_is_good'])
        
        df[''] += df[normalized_col_name] * config['weight']

    df[''] *= 100

    df_sorted = df.sort_values(by='', ascending=False)

    # ---  ---
    # 2
    display_columns = [
        'substrate', 'material', 'pressure_GPa', 'high_temp_K', '',
        'tensile_strength_GPa',                # 
        'interfacial_contamination_density',     #  ()
        'substrate_contamination_density',     #  ()
        'cathode_contamination_density',       #  ()
    ]
    
    display_columns = [col for col in display_columns if col in df_sorted.columns]

    print("\n--- 10 () ---")
    print(df_sorted[display_columns].head(10).round(4)) # 
    print("\n--- 10 () ---")
    print(df_sorted[display_columns].tail(10).round(4))

    output_filename = "output/comprehensive_analysis_output/interfacial_density_scored_results_detailed.csv"
    df_sorted.to_csv(output_filename, index=False)
    print(f"\n '{output_filename}' ")

if __name__ == "__main__":
    score_for_clean_separation()
```

```python
#
import pandas as pd
import numpy as np

def calculate_clean_separation_score(df):
    """
    Excel
    0
    """
    mixing_scores = []
    substrate_ratios = []
    cathode_ratios = []
    # DataFrame
    for index, row in df.iterrows():
        # --- 1.  ---
        substrate_name = row.get('substrate')
        substrate_elements = []
        if substrate_name == 'Al':
            substrate_elements = ['Al']
        elif substrate_name == 'Al2O3':
            # 'O''Al'
            substrate_elements = ['Al']
        elif substrate_name == 'AlF3':
            substrate_elements = ['Al', 'F']

        substrate_upper_sum = 0
        substrate_total_sum = 0
        for elem in substrate_elements:
            upper_val = row.get(f'{elem}_upper', 0)
            lower_val = row.get(f'{elem}_lower', 0)
            substrate_upper_sum += upper_val
            substrate_total_sum += (upper_val + lower_val)

        if substrate_total_sum > 0:
            substrate_mixing_ratio = substrate_upper_sum / substrate_total_sum
        else:
            substrate_mixing_ratio = 1.0  # 

        # --- 2.  ---
        cathode_elements = ['Li', 'Mn', 'Co', 'Ni']
        cathode_lower_sum = 0
        cathode_total_sum = 0
        for elem in cathode_elements:
            upper_val = row.get(f'{elem}_upper', 0)
            lower_val = row.get(f'{elem}_lower', 0)
            cathode_lower_sum += lower_val
            cathode_total_sum += (upper_val + lower_val)
        
        if cathode_total_sum > 0:
            cathode_mixing_ratio = cathode_lower_sum / cathode_total_sum
        else:
            cathode_mixing_ratio = 1.0 # 
        
        # --- 3.  ---
        if substrate_total_sum == 0 and cathode_total_sum == 0:
            mixing_score = 2.0
            substrate_ratios.append(1.0)
            cathode_ratios.append(1.0)
        else:
            mixing_score = substrate_mixing_ratio + cathode_mixing_ratio
            substrate_ratios.append(substrate_mixing_ratio)
            cathode_ratios.append(cathode_mixing_ratio)
        
        mixing_scores.append(mixing_score)

    df['clean_separation_score'] = mixing_scores
    df['substrate_mixing_ratio'] = substrate_ratios
    df['cathode_mixing_ratio'] = cathode_ratios
    return df

def normalize_metric(series, high_is_good=True):
    """
    PandasSeries
    """
    series_finite = series.replace([np.inf, -np.inf], np.nan).fillna(0)
    min_val = series_finite.min()
    max_val = series_finite.max()
    
    if max_val - min_val == 0:
        return pd.Series(0.5, index=series.index)

    if high_is_good:
        return (series_finite - min_val) / (max_val - min_val)
    else:
        return (max_val - series_finite) / (max_val - min_val)

def score_for_clean_separation(csv_path="/home/jovyan/Kaori/MD/LiB_2/structure/comprehensive_analysis_output/comprehensive_analysis_results_2.csv"):
    """
    
    """
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f":  '{csv_path}'")
        return

    print("...")

    # --- 1:  ---
    metrics_config = {
        'tensile_strength_GPa': {'weight': 0.1, 'high_is_good': False},
        'fracture_step':        {'weight': 0.0, 'high_is_good': False},
        'substrate_mixing_ratio': {'weight': 0.45, 'high_is_good': False},
        'cathode_mixing_ratio':   {'weight': 0.45, 'high_is_good': False},
    }

    # --- 2:  ---
    df = calculate_clean_separation_score(df)

    # 
    # --- 3:  ---
    # 
    # 0
    print("\nlog...")
    transform_targets = ['substrate_mixing_ratio', 'cathode_mixing_ratio']
    for col in transform_targets:
        if col in df.columns:
            # 
            new_col_name = f'before_{col}'
            # 2: 
            df[new_col_name] = df[col].copy()
            df[col] = np.log1p(df[col])
            print(f"- '{col}' ")
    # 
    # ---  ---
    # 

    # --- 4:  ---
    df[''] = 0.0
    
    print("\n:")
    for metric, config in metrics_config.items():
        direction = "" if not config['high_is_good'] else ""
        print(f"- {metric}:  {config['weight']*100:.0f}%, ({direction})")
        
        df[metric].fillna(0, inplace=True)
        
        # 
        normalized_col_name = f'norm_{metric}'
        df[normalized_col_name] = normalize_metric(df[metric], high_is_good=config['high_is_good'])
        
        # 
        df[''] += df[normalized_col_name] * config['weight']

    df[''] *= 100

    df_sorted = df.sort_values(by='', ascending=False)

    # ---  ---
    display_columns = [
        'substrate', 'material', 'pressure_GPa', 'high_temp_K', ''
    ] + list(metrics_config.keys())

    print("\n--- 10 () ---")
    print(df_sorted[display_columns].head(10).round(5))

    print("\n--- 10 () ---")
    print(df_sorted[display_columns].tail(10).round(5))
    output_filename = "./output/comprehensive_analysis_output/log_scored_results_for_clean_separationcsv"
    df_sorted.to_csv(output_filename, index=False)
    print(f"\n '{output_filename}' ")

if __name__ == "__main__":
    score_for_clean_separation()
```

```python
# create_initial_state_details_csv.py

from ase.io import Trajectory,read
import os
import numpy as np
import pandas as pd
from collections import Counter # 

# ---  ---
# 
ATOMS_LIST = ["Li3MnCoNiO6", "NiO", "Li2O", "Mn3O4", "Co3O4"]
COMPOUNDS = ["Al", "Al2O3", "AlF3"]

# 
COMPRESSION_TEMPERATURES = [300]
COMPRESSION_PRESSURES = [0.001, 0.005, 0.01]
HIGH_TEMPERATURES = [300]
TENSILE_TEMPERATURE = 300.0

# !!! :  !!!
TENSILE_DIR = "/home/jovyan/Kaori/MD/LiB_2/strucure/MD/tensile"  # : "/home/jovyan/Kaori/MD/LiB_2/structure/slow01/tensile"
INPUT_DIR = "/home/jovyan/Kaori/MD/LiB_2/strucutre"
# ---  ---
# straction
UPPER_LAYER_SYMBOLS = {'Li', 'Mn', 'Co', 'Ni'}
def get_input_file_path(com, atom):
    """"""
    return os.path.join(INPUT_DIR, f"{com}_{atom}_interface.xyz")

def get_input_file_traj(com, atom):
    """"""
    return os.path.join(INPUT_DIR, f"{com}_{atom}_interface_optimized.traj")
    
def get_tensile_file_path(sub, base_name, pressure, comp_temp, high_temp, tensile_temp):
    """"""
    return os.path.join(
        TENSILE_DIR,
        f"{sub}_{base_name}_P{pressure:.3f}_T{comp_temp:.0f}K_HT{high_temp}K_tensile{tensile_temp:.0f}K_early_stop.traj"
    )

def calculate_and_save_initial_details():
    """
    
    ZCSV
    """
    print("=== CSV ===")
    
    results_list = []

    for sub in COMPOUNDS:
        for base_name in ATOMS_LIST:
            for pressure in COMPRESSION_PRESSURES:
                for comp_temp in COMPRESSION_TEMPERATURES:
                    for high_temp in HIGH_TEMPERATURES:
                        
                        traj_path = get_input_file_path(sub, base_name)

                        if not os.path.exists(traj_path):
                            continue

                        try:
                            traj = read(traj_path)
                            if len(traj) == 0:
                                continue
                            
                            initial_atoms = traj[0]
                            
                            # --- 1.  ---
                            symbols = initial_atoms.get_chemical_symbols()
                            atom_counts = Counter(symbols) # : {'Li': 24, 'O': 96, ...}
                            
                            # --- 2. Z ---
                            positions = initial_atoms.get_positions()
                            
                            # 
                            upper_layer_indices = [i for i, symbol in enumerate(symbols) if symbol in UPPER_LAYER_SYMBOLS]
                            
                            upper_layer_z_width = 0.0
                            if upper_layer_indices:
                                # Z
                                upper_z_coords = positions[upper_layer_indices, 2]
                                upper_layer_z_width = np.max(upper_z_coords) - np.min(upper_z_coords)

                            # ---  ---
                            row = {
                                'substrate': sub,
                                'material': base_name,
                                'pressure_GPa': pressure,
                                'high_temp_K': high_temp,
                                'upper_layer_z_width_A': upper_layer_z_width
                            }
                            
                            # : 'count_Li', 'count_O'
                            for symbol, count in atom_counts.items():
                                row[f'count_{symbol}'] = count
                            
                            results_list.append(row)

                        except Exception as e:
                            print(f": {traj_path} - {e}")
                            continue
    
    if not results_list:
        print("CSV")
        return

    # Pandas DataFrame
    df = pd.DataFrame(results_list)

    # DataFrame
    # 
    id_cols = ['substrate', 'material', 'pressure_GPa', 'high_temp_K']
    # 
    data_cols = ['upper_layer_z_width_A']
    # 
    count_cols = sorted([col for col in df.columns if col.startswith('count_')])
    
    # DataFrame
    df = df[id_cols + data_cols + count_cols]

    # NaN0
    df.fillna(0, inplace=True)
    # 
    for col in count_cols:
        df[col] = df[col].astype(int)

    # CSV
    output_filename = "input_initial_state_details.csv"
    df.to_csv(output_filename, index=False, float_format='%.4f')

    print(f"\n===  ===")
    print(f" '{output_filename}' ")
    print(f": {len(df)}")


if __name__ == "__main__":
    # ASE, NumPy, Pandas
    # pip install ase numpy pandas
    calculate_and_save_initial_details()
```

```python
# ############################################################################
# 
# ############################################################################

def calculate_mixing_atom_count(df):
    """
    
    -  (Al, F)
    -  (Li, Mn, Co, Ni)
    2
    """
    # --- ---
    substrate_elements = ['Al']
    df['substrate_mixed_atoms'] = 0
    for elem in substrate_elements:
        col = f'{elem}_upper'
        if col in df.columns:
            df['substrate_mixed_atoms'] += df[col].fillna(0)

    # --- ---
    cathode_elements = ['Li', 'Mn', 'Co', 'Ni']
    df['cathode_mixed_atoms'] = 0
    for elem in cathode_elements:
        col = f'{elem}_lower'
        if col in df.columns:
            df['cathode_mixed_atoms'] += df[col].fillna(0)

    # ---  ---
    df['total_mixed_atoms'] = df['substrate_mixed_atoms'] + df['cathode_mixed_atoms']
    return df

def normalize_metric(series, high_is_good=True):
    """
    PandasSeries
    high_is_goodFalse
    """
    series_finite = series.replace([np.inf, -np.inf], np.nan).fillna(0)
    min_val = series_finite.min()
    max_val = series_finite.max()
    
    if max_val - min_val == 0:
        return pd.Series(0.5, index=series.index) # 0.5

    if high_is_good:
        #  ()
        return (series_finite - min_val) / (max_val - min_val)
    else:
        #  ()
        return (max_val - series_finite) / (max_val - min_val)

def score_for_clean_separation(csv_path="/home/jovyan/Kaori/MD/LiB_2/strucure/comprehensive_analysis_output/comprehensive_analysis_results_2.csv"):
    """
    
    
    """
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f":  '{csv_path}'")
        return

    print("...")

    # --- 1:  ---
    # 'high_is_good': TrueFalse
    # total_mixed_atoms ()
    metrics_config = {
        'tensile_strength_GPa': {'weight': 0.2, 'high_is_good': False},
        'fracture_step':        {'weight': 0.0, 'high_is_good': False},
        'total_mixed_atoms':    {'weight': 0.8, 'high_is_good': False}, # : 
    }

    # 
    df = calculate_mixing_atom_count(df) # : 

    # ---  ---
    df[''] = 0.0
    
    print("\n:")
    for metric, config in metrics_config.items():
        direction = "" if not config['high_is_good'] else ""
        print(f"- {metric}:  {config['weight']*100:.0f}%, ({direction})")
        
        # 
        if metric not in df.columns:
            print(f"  :  '{metric}' ")
            continue
            
        df[metric].fillna(0, inplace=True)
        
        #  ()
        normalized_col_name = f'norm_{metric}'
        df[normalized_col_name] = normalize_metric(df[metric], high_is_good=config['high_is_good'])
        
        # 
        df[''] += df[normalized_col_name] * config['weight']

    df[''] *= 100 # 100

    # 
    df_sorted = df.sort_values(by='', ascending=False)

    # ---  ---
    display_columns = [
        'substrate', 'material', 'pressure_GPa', 'high_temp_K', ''
    ] + list(metrics_config.keys())
    
    # 
    display_columns = [col for col in display_columns if col in df_sorted.columns]

    print("\n--- 10 () ---")
    print(df_sorted[display_columns].head(10).round(2))
    print("\n--- 10 () ---")
    print(df_sorted[display_columns].tail(10).round(2))

    # 
    output_filename = "output/comprehensive_analysis_output/atom_count_scored_resultst.csv"
    df_sorted.to_csv(output_filename, index=False)
    print(f"\n '{output_filename}' ")

if __name__ == "__main__":
    # run_comprehensive_analysis() # 
    score_for_clean_separation()   # 
```
